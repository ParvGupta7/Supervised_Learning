```python
#Given a function f and a starting point w(weight),
#try to find the minima or gradient descent
#(when we will call the function, it will keep on oscillating)

def gradient_descent(func,w): #here w is initial weight
  list_of_weights = [] # we will append weight after every iteration to it.
  weight = w #lets say w = 10.
  delta = 0.0001
  for i in range(100):
    derivative = (func(weight+delta) - func(weight))/delta  # as per formula (del f/del w)
    weight = weight - derivative  # as per formula, w1 = w1 - derivative
    list_of_weights.append(weight)
  return list_of_weights

# It will keep on oscillating between -14 and 9.9 cuz there is no learning rate
gradient_descent(lambda x: x**2 + 4*x + 3, 10)





```

```python
# We will multiply our derivative with the learning rate to stop it from oscillations

def gradient_descent(func,w): #here w is initial weight
  list_of_weights = [] # we will append weight after every iteration to it.
  weight = w #lets say w = 10.
  delta = 0.0001
  learning_rate = 0.1 #LEARNING RATE
  for i in range(1000): #changed to 1000 iterations
    derivative = (func(weight+delta) - func(weight))/delta  # as per formula (del f/del w)
    weight = weight - (learning_rate * derivative)  # as per formula, w1 = w1 - derivative
    list_of_weights.append(weight)
  return list_of_weights

# Now the weights would decrease and stabilised to -2.0, its the minima of loss function, it is the point where the error would be minimum.
gradient_descent(lambda x: x**2 + 4*x + 3, 10)
```
